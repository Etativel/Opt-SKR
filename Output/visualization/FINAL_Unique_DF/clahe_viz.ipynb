# CLAHE 0.1

```python
import pandas as pd

df_clahe = pd.read_csv("clahe_unique_final_withmodel.csv")
df_clahe
```

```python
import matplotlib.pyplot as plt

# Sample data
models = df_clahe['gen']
accuracy = df_clahe['acc']

# Create line chart
plt.figure(figsize=(10, 6))
plt.plot(models, accuracy, marker='o', linestyle='-', color='black')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.grid(True)
plt.ylim(0.95, 1.00)

# Customize for scientific presentation
plt.xticks(models, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Save the plot as an image file
plt.savefig('Model_Accuracy_Comparison.png')
plt.show()
```
