```python
import matplotlib.pyplot as plt

# Assuming df_corrected is already sorted by generation if needed
plt.figure(figsize=(12, 6))  # Adjusted for a more proportionate display

# Setting a style
plt.style.use('seaborn-dark')  # Elegant and understated background

# Plotting the data
plt.plot(df_corrected['gen'], df_corrected['acc'], linestyle='-', linewidth=2, color='deepskyblue')  # Smooth and professional color choice

# Title and labels with enhanced font properties
plt.title('Accuracy by Generation', fontsize=16, fontweight='bold', color='navy')
plt.xlabel('Generation', fontsize=12, fontweight='regular', color='navy')
plt.ylabel('Accuracy', fontsize=12, fontweight='regular', color='navy')

# Customizing tick parameters for better readability
plt.xticks(fontsize=10, fontweight='bold', color='gray')
plt.yticks(fontsize=10, fontweight='bold', color='gray')

# Enhancing the grid for better precision viewing
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Display the plot with a clean and professional layout
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate a simple moving average with a window size of your choice
window_size = 5  # Change this based on your data size and desired smoothing
df_corrected['smoothed_accuracy'] = df_corrected['acc'].rolling(window=window_size, center=True).mean()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df_corrected['gen'], df_corrected['smoothed_accuracy'], linestyle='-', linewidth=2)
plt.title('Smoothed Accuracy by Generation')
plt.xlabel('Generation')
plt.ylabel('Smoothed Accuracy')
plt.grid(True)
plt.show()
```
