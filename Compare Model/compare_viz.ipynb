# Normal

```python
import pandas as pd

df_normal1 = pd.read_csv("normal_trial1.csv")
df_normal1
```

```python
import matplotlib.pyplot as plt

# Sample data
models = df_normal1['iteration']
accuracy = df_normal1['acc']

# Create line chart
plt.figure(figsize=(10, 6))
plt.plot(models, accuracy, marker='o', linestyle='-', color='black')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Model Consistency')
plt.grid(True)
plt.ylim(0.95, 1.00)

# Customize for scientific presentation
plt.xticks(models, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Save the plot as an image file
plt.savefig('Model_Accuracy_Comparison.png')
plt.show()
```
